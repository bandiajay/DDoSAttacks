```python
myData =  myData.replace([np.inf, -np.inf], np.nan)
myData = myData.fillna(0)
```

```python
from sklearn.model_selection import train_test_split
X = myData.drop('Label', axis = 1)
Y = myData['Label']
Xtrain_set, Xtest_set,Ytrain_set, Ytest_set  = train_test_split(X,Y,test_size = 0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import matplotlib.pyplot as plt
error_rate = []
for i in range(1,600):
 knn = KNeighborsClassifier(n_neighbors=i)
 knn.fit(Xtrain_set,Ytrain_set)
 pred_i = knn.predict(Xtest_set)
 error_rate.append(np.mean(pred_i != Ytest_set))

plt.figure(figsize=(10,6))
plt.plot(range(1,600),error_rate,color='blue', linestyle='dashed', 
         marker='o',markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
print("Minimum error:-",min(error_rate),"at K =",error_rate.index(min(error_rate)))
```
